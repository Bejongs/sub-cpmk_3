```python
import re
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/cases.csv")

print(df.shape)
df.head()
```

```python
def extract_no_perkara(text):
    pattern = r'(?:nomor\s*)?(\d+/pdt\.g/\d{4}/pn\s*sda)'
    match = re.search(pattern, text.lower())
    return match.group(1).replace(" ", "") if match else "-"

def extract_penggugat(text):
    text = text.lower()
    match = re.search(r'antara\s+(.*?)\s+lawan', text)
    return match.group(1)[:500] if match else "-"

def extract_tergugat(text):
    text = text.lower()
    match = re.search(r'lawan\s+(.*?)(?:menimbang|setelah membaca|dalam hal ini)', text)
    return match.group(1)[:500] if match else "-"

def extract_amar(text):
    text = text.lower()
    
    if "mengadili" in text:
        return text.split("mengadili")[-1][:1000]
    
    if "menetapkan" in text:
        return text.split("menetapkan")[-1][:1000]
    
    return "-"
```

```python
df["no_perkara"] = df["text_full"].apply(extract_no_perkara)
df["penggugat"] = df["text_full"].apply(extract_penggugat)
df["tergugat"] = df["text_full"].apply(extract_tergugat)
df["amar_putusan"] = df["text_full"].apply(extract_amar)
df["jenis_perkara"] = "Wanprestasi"
df["pengadilan"] = "PN Sidoarjo"

df.head()
```

```python
df.to_csv("../data/processed/cases_final.csv", index=False)

print("cases_final.csv berhasil dibuat")
print(df.shape)
```
